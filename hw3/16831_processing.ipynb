```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# DQN

# Step 1: Load the CSV file into a DataFrame
df_dqn = pd.read_csv('dqn.csv', header=None)

# Step 2: Display the first few rows of the DataFrame
df_dqn.head()
```

```python
# DDQN

# Step 1: Load the CSV file into a DataFrame
df_ddqn = pd.read_csv('ddqn.csv', header=None)

# Step 2: Display the first few rows of the DataFrame
df_ddqn.head()
```

```python

# Step 2: Calculate the average and standard deviation for each experiment
df_dqn['Average'] = df_dqn[[1, 2, 3]].mean(axis=1)
df_dqn['Std_Dev'] = df_dqn[[1, 2, 3]].std(axis=1)

# Step 2: Calculate the average and standard deviation for each experiment
df_ddqn['Average'] = df_ddqn[[1, 2, 3]].mean(axis=1)
df_ddqn['Std_Dev'] = df_ddqn[[1, 2, 3]].std(axis=1)

# Step 3: Create the line plot with error bars
plt.figure(figsize=(10, 6))

# Assuming 'Steps' is the name of the first column
plt.errorbar(df_dqn[0], df_dqn['Average'], yerr=df_dqn['Std_Dev'], fmt='o-', label='DQN')
# Assuming 'Steps' is the name of the first column
plt.errorbar(df_ddqn[0], df_ddqn['Average'], yerr=df_ddqn['Std_Dev'], fmt='o-', label='DDQN')


# Adding labels and title
plt.xlabel('Steps')
plt.ylabel('Average Result')
plt.title('Average Result with Error Bars')
plt.legend()
# Display the plot
plt.show()
```
