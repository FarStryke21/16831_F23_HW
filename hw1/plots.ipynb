```python
# Plot of Returns when changing Learning Rate
import matplotlib.pyplot as plt

# Assuming x_values, y_values, and std_dev are your lists
x_values = [1e-3, 2e-3, 3e-3, 4e-3, 5e-3]
y_values = [207.4, 210.38, 458.31, 977.04, 1138.81]
std_dev = [2.18, 1.78, 3.22, 54.38, 45.39]

# Plotting the data points
plt.plot(x_values, y_values, 'b-', label='Returns')

# Adding error bars for standard deviation
plt.errorbar(x_values, y_values, yerr=std_dev, fmt='none', ecolor='red', elinewidth=1, capsize=4, label='Standard Deviation')

# Labeling the axes and adding legend
plt.xlabel('Learning Rate')
plt.ylabel('Average Return of the Hopper Policy')
plt.legend()

# Show the plot
plt.show()
```
